```python
from bs4 import BeautifulSoup
import requests
```

```python
url = "https://books.toscrape.com/"
```

```python
page = requests.get(url)
```

```python
books
```

```python
book_data = []
```

```python
soup.find_all("article", class_="product_pod")
```

```python
Articles = soup.find_all("article", class_="product_pod")
```

```python
Articles
```

```python
for a in Articles:
    title = a.h3.a.get_text(strip=True) 
    price = a.find('p',class_="price_color").get_text()
    stock = a.find('p',class_="instock availability").get_text(strip=True)
    book_data.append({
        "title":title,
        "price":price,
        "availability":stock
    })
```

```python
book_data
```

```python
for page in range(2, 51):  # there are 50 pages
    url = f"http://books.toscrape.com/catalogue/page-{page}.html"
    response = requests.get(url)
    
    if response.status_code != 200:
        print(f"⚠️ Skipping page {page}, status code: {response.status_code}")
        continue

    soup = BeautifulSoup(response.text, "lxml")
    articles = soup.find_all("article", class_="product_pod")

    for a in articles:
        title = a.h3.a.get("title") or a.h3.a.get_text(strip=True)
        price = a.find("p", class_="price_color").get_text(strip=True)
        availability = a.find("p", class_="instock availability").get_text(strip=True)

        book_data.append({
            "title": title,
            "price": price,
            "availability": availability
        })

    print(f"✅ Scraped page {page}, total books so far: {len(book_data)}")

print("🎉 Finished scraping! Total books collected:", len(book_data))
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(book_data)
```

```python
df
```

```python
df.to_csv("books_raw.csv", index=False, encoding="utf-8")
```
